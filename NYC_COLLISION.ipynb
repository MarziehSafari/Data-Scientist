```python
import pandas
df1 = pandas.read_csv("https://data.cityofnewyork.us/resource/h9gi-nx95.csv")
df1.to_sql("NYC_CRASHES", con, if_exists='replace', index=False, method="multi")

df2 = pandas.read_csv("https://data.cityofnewyork.us/resource/bm4k-52h4.csv")
df2.to_sql("NYC_VEHICLES", con, if_exists='replace', index=False, method="multi")

df3 = pandas.read_csv("https://data.cityofnewyork.us/resource/f55k-p6yu.csv")
df3.to_sql("NYC_PERSONS", con, if_exists='replace', index=False, method="multi")
```
